##### 1. Generate a 1D array of 1000 random numbers using NumPy from a normal distribution and
##### compute statistics.
###### (i) Compute the mean of the data
###### (ii) Compute the standard deviation
###### (iii) Find the maximum and minimum values
###### (iv) Count how many values are above the mean

```python
import numpy as np
data = np.random.randn(1000)
mean_val = np.mean(data)
print("Mean value",mean_val)
std_val = np.std(data)
print("standard deviatio",std_val)
maxi = np.max(data)
print("Maximum value:",maxi)
mini = np.min(data)
print("Minimum value",mini)
above_mean = np.sum(data>mean_val)
print("Number of values above mean is ",above_mean)
```

##### 2. Filter elements from a NumPy array based on the given conditions.
###### (i) Select all even numbers
###### (ii) Select numbers divisible by 3 or 5
###### (iii) Replace all numbers greater than 40 with -1

```python
arr = np.random.randint(1, 100, 50)
print(arr)
even_number = arr[arr %2 == 0]
print(even_number)
div_by_3_or_5 = arr[(arr %3 ==0) | (arr%5==0)]
print(div_by_3_or_5)
modified_arr = np.where(arr>40,-1,arr)
print(modified_arr)
```

HINT: Use df.head(), df.describe(), conditional column creation, and df.sort_values()

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, 25, 22, 24, 23],
    'Grade': [88, 92, 85, 95, 90]
}
df = pd.DataFrame(data)
print(df.head(3))
print(df.describe())
df["Passed"] = df["Grade"]>=90
print(df)
sorted_df = df.sort_values(by = "Grade",ascending=False)
print(sorted_df)
```
