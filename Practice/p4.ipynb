```python
top_10=data.sort_values("Runtime (Minutes)", ascending=False)[["Title", "Runtime (Minutes)"]].head(10)
```

```python
# top_10=data.sort_values("Runtime (Minutes)", ascending=False)[["Title", "Runtime (Minutes)"]].head(10)
top10_len=data.nlargest(10,"Runtime (Minutes)")[["Title","Runtime (Minutes)"]].set_index("Title")
plt.title("Top 10 Longest Movies by Runtime")
plt.xlabel("Runtime (Minutes)")
plt.ylabel("Movie Title")
sns.barplot(x="Runtime (Minutes)", y=top10_len.index, data=top10_len, palette="viridis")
plt.show()
```
