# Matplotlib functions

```python
#plt.plot(x,y,color="color name",linestyle="line style",linewidth=value,marker="marker symbol",label="label name")

months=[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.plot(months,sales,color="blue",linestyle="--",linewidth=2,marker="o",label="2025 sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly sales data report")
# plt.legend(loc="upper left")
plt.legend(loc="lower right")
plt.grid(color="pink",linewidth=1,linestyle=":")
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],["m1","m2","m3","m4"])
plt.show()
```

# Bar Graph

```python
#plt.bar(x,height,color="color name",width = value,label="label name")
product = ["A","B","C","D"]
sales=[1000,1500,800,1200]
plt.bar(product,sales,color="orange",label="sales 2025")
plt.xlabel("product")
plt.ylabel("Sales")
plt.title("Product sale comparison")
plt.legend()
plt.show()
```

# Pie Chart

```python
#plt.pie(values,labels=label_list,colors=color_list,autopct="%1.1f%%")
regions = ["North","South","East","West"]
revenue = [3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct="%1.1f%%",colors=["gold","skyblue","lightgreen","coral"])
plt.title("Revenue contribution by region")
plt.show()
```

# Histograms

```python
#plt.hist(data,bins=number_of_bins,color="color name",edgecolor="color name")
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]
plt.hist(scores,bins=5,color="purple",edgecolor="black")
plt.xlabel("Score range")
plt.ylabel("Number of students")
plt.title("Score distribution of students")
plt.show()
```

# Scatter Plots

```python
#plt.scatter(x,y,color="color name",marker="marker style",label="label name")
hours_studied=[1,2,3,4,5,6,7,8]
exam_scores=[50,55,60,65,70,75,80,85]
plt.scatter(hours_studied,exam_scores,color="green",marker="o",label="student data")
plt.xlabel("Hours_studied")
plt.ylabel("Exam_scores")
plt.title("Relationship between study time and scores")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
#how to compare two classes data

# hours_studied=[1,2,3,4,5,6,7,8]
# exam_scores=[50,55,60,65,70,75,80,85]
plt.scatter([1,2,3],[50,55,60],color="green",marker="o",label="class A")#gp1
plt.scatter([1,2,3],[45,50,52],color="orange",marker="o",label="class B")#gp2
plt.xlabel("Hours_studied")
plt.ylabel("Exam_scores")
plt.title("Comparison of two classes")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1) # 1 row,2column,1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2) # 1 row,2column,2nd subplot
plt.bar(x,y)
plt.title("Bar Chart")
plt.tight_layout()
plt.show()
```

```python
#professional way of creating uppapar wala way bhi shi hai lekin can use any

x=[1,2,3,4]
y=[10,20,15,25]

#fig, ax=plt.subplots(nrows,ncols,figsize=(width,height))
fig,ax = plt.subplots(1,2,figsize = (10,5))
ax[0].plot(x,y,color="blue")
ax[0].set_title("Line Plot")
ax[1].bar(x,y,color="green")
ax[1].set_title("Bar_Chart")
fig.suptitle("Comparison of line and bar charts")
plt.tight_layout()
plt.show()
```
